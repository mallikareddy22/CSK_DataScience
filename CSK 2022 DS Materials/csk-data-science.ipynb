```python
#We are using a Pandas method to read in a csv file as a dataFrame
women_stem = pd.read_csv('women-stem.csv')
women_stem
```

```python
#The where method takes in a conditional and only keeps rows where the conditional is True and NaN for other rows
#The following method will return a table only with rows where the Major category is engineering and the amount of Women is less than 500

#To get column values: table[column name]
#& evaluates to and

women_stem.where((women_stem["Major_category"] == "Engineering") & (women_stem["Women"] < 500))
```

```python
#To get rid of the null values, we use the dropna() function
women_stem.where((women_stem["Major_category"] == "Engineering") & (women_stem["Women"] < 500)).dropna()
#Notice how the table below and above differ
```

```python
#Exercise 1
#Return the number of Engineering majors where ShareWomen is less than 0.5.
#https://datagy.io/pandas-number-of-rows/ is a great material that can help if you are confused.

low_engineering = women_stem.where((women_stem["Major_category"] == "Engineering") & (women_stem["ShareWomen"] < 0.5)).dropna()
num_low_engineering = len(low_engineering.index)
num_low_engineering
```

```python
#Exercise 2
#Using your previous answer, what is the percentage of Engineering majors where ShareWomen is less than 0.5 compared to all Engineering majors

engineering_majors = women_stem.where(women_stem["Major_category"] == "Engineering").dropna()
engineering_majors_length = len(engineering_majors.index)
percentage_women = num_low_engineering/engineering_majors_length * 100
percentage_women
```

```python
#Shocking number, right? In all the engineering majors on this table, women are only less than 50 percent (consistently the minority)
#Let's look into why the amount of women in Engineering majors is so low and how we can help improve this ratio.
#Some reasons include a lack of role models and Sexism in STEM Departments and workplaces.
#We can improve this through highlighting women scientists, engaging young girls through programs like CSK, and fighting sexism in STEM.
#Read more here: https://www.wgu.edu/blog/why-are-there-so-few-women-in-stem1907.html#close
```

```python
#Let's figure out how to find the major with the highest amount of women.
#To do this, we need to learn about the sort method
#We are first going to sort the "Women" column in descending order
#See the documentation of sort_values here: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_values.html
# When confused about what the parameters of a function are or what it returns, look up the documentation of the function just like above.
sorted_table = women_stem.sort_values(by=['Women'], ascending=False)
#To return the first row we can use the head() function which returns the first five rows of a dataframe and pass in 1 as the parameter to get the first row.
first_row_sorted = sorted_table.head(1)
#To get the Major of the first row we need to see what value is in the Major column
first_row_sorted["Major"]
```

```python
#Your turn!
#Exercise 3
#What are the five majors that have the lowest proportions of women.
#Hint: We are looking at the "ShareWomen" column and try utilizing the functions from above.
sorted_women = women_stem.sort_values(by=['ShareWomen'], ascending=True)
five_rows_sorted = sorted_women.head()
five_rows_sorted["Major"]
```

```python
#Exercise 4
#What is the engineering major with the highest number of men?
#Hint: You will need to use the where function and the functions from above.
engineering_majors = women_stem.where(women_stem["Major_category"] == "Engineering")
engineering_majors_sorted = engineering_majors.sort_values(by=["Men"], ascending=False)
engineering_majors_sorted_firstrow = engineering_majors_sorted.head(1)
highest_number_men = engineering_majors_sorted_firstrow["Major"]
highest_number_men
```

```python
#Next, let's try grouping our data
#groupby groups the table rows by a certain column name and returns a groupby object
grouped_women_stem = women_stem.groupby('Major_category')

#There are certain methods we can use on groupby objects
#count() counts the number of rows in each group
#For example, there are 14 rows in the major category of Biology & Life Science
grouped_women_stem.count()
```

```python
#Take a look at this website and try out some more groupby methods on your own
#https://towardsdatascience.com/all-pandas-groupby-you-should-know-for-grouping-data-and-performing-operations-2a8ec1327b5
grouped_women_stem.get_group("Engineering")
#This returns all the rows in the Engineering group of the Major category
```

```python
#Exercise 5
#What is the maximum number of women in the Physical Sciences major category? There are multiple ways to do this, but try doing this with groupby!
women_stem_grouped = women_stem.groupby("Major_category")
women_stem_physical_sciences = women_stem_grouped.get_group("Physical Sciences")
physical_sciences_women = women_stem_physical_sciences["Women"]
physical_sciences_women_max = physical_sciences_women.max()
physical_sciences_women_max
```

```python
#Exercise 6
#What is the minimum ShareWomen percentage for Majors with a Median of 50000?
group_median = women_stem.groupby("Median")
group_median_50000 = group_median.get_group(50000)
group_median_50000_sharewomen = group_median_50000["ShareWomen"]
min_sharewomen_50000 = min(group_median_50000_sharewomen)
min_sharewomen_50000
```

```python
#Let's now explore how to plot our data for more analysis
physical_majors = women_stem.where(women_stem["Major_category"] == "Physical Sciences").dropna()
#When we want to compare some value across some subgroups we use bar charts.
#To create any kind of plot (ex. bar, line, scatter, histogram, etc.), we use the plot method, specifying the x and y axes and the kind of plot.
#Below visualizes a bar graph comparing the amount of women in different Physical Sciences majors.
physical_majors.plot(x="Major", y="Women", kind="bar")
```

```python
#When we want to explore the relationship between two variables, we use a scatter plot.
#Let's say that we want to see if there is a relationship between the number of Women and the number of Men
physical_majors.plot(x="Men", y="Women", kind="scatter")
#Do you think there is some kind of relationship between these two variables? 
```

```python
#Exercise 7
#Plot the mean ShareWomen percentages in a bar graph for each Major category.
#Hint: Use groupby!
women_stem.groupby("Major_category")["ShareWomen"].mean().plot(kind="bar")
#What do you notice?
```

```python
#Exercise 8
#Plot the relationship between the Median (x) and the ShareWomen percentages (y)in a scatter plot.
women_stem.plot(x="Median", y="ShareWomen", kind="scatter")
```

```python
#Exercise 9
#Print out the major with the smallest ShareWomen value and create a bar graph showing the number of men for each major in the Health major category.
sorted_sharewomen = women_stem.sort_values(by=['ShareWomen'], ascending=True)
first_row_sorted = sorted_sharewomen.head(1)
major_smallest = first_row_sorted["Major"]
print(major_smallest)
health_majors = women_stem.where(women_stem["Major_category"] == "Health").dropna()
health_majors.plot(x="Major", y="Men", kind="bar")
```
